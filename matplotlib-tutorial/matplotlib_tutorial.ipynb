# Data Visualization Tutorial

#### Import necessary libraries 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Line graph

```python
# initialize coordinates
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# resize the graph
# adjust pixels per inch (dpi)
plt.figure(figsize = (10, 7), dpi = 100)

# initialize, label, and color graph
# adjust width of line
# add markers to coordinate pairs on graph 
# adjust marker size
plt.plot(x, y, label = '2x', color = 'green', linewidth = 2, marker = '.', markersize = 10) 

# make second line on graph
# numpy allows us to plot x-values ranging from 0 to 4.5, incrementing by 0.5
# y value is x-value squared
# make color red and label line x^2 
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, color = 'r', label = 'x^2')

# title the graph
# change font style and font size of title
plt.title('Line Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20}) 

plt.xlabel('x-axis') # label the x-axis
plt.ylabel('y-axis') # label the y-axis

plt.xticks([0, 1, 2, 3, 4]) # only show integer values on x-axis
plt.yticks([0, 2, 4, 6, 8]) # only show integer values on y-axis

plt.legend() # show label

# save the graph
plt.savefig('linegraph.png', dpi = 300)

plt.show() # show plot
```

#### Bar chart

```python
# label bars on graph
labels = ['A', 'B', 'C']

# initialize values 
values = [2, 4, 6]

# resize the graph
# adjust pixels per inch (dpi)
plt.figure(figsize = (10, 7), dpi = 100)

# create bar chart
bars = plt.bar(labels, values)

# add effects to the bars 
# pop with no argument removes the last item in list first
# pop with argument 0 removes the first item in list first

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# show bar chart
plt.show()
```
